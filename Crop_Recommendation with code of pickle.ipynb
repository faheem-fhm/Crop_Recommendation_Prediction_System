```python
df['Nitrogen'].fillna(df['Nitrogen'].mean(),inplace=True)
print(df['Nitrogen'].head())
df.isnull().sum()
```

```python
df['Phosphorus'].fillna(df['Phosphorus'].median(),inplace=True)
print(df['Phosphorus'].head())
df.isnull().sum()
```

```python
df['Potassium'].fillna(df['Potassium'].median(),inplace=True)
print(df['Potassium'].head())
df.isnull().sum()
```

```python
df['Humidity'].fillna(df['Humidity'].mean(),inplace=True)
print(df['Humidity'].head())
df.isnull().sum()
```

```python
df['pH_Value'].fillna(df['pH_Value'].median(),inplace=True)
print(df['pH_Value'].head())
df.isnull().sum()
```

```python
df['Rainfall'].fillna(df['Rainfall'].median(),inplace=True)
print(df['Rainfall'].head())
df.isnull().sum()
```

```python
df['Crop'].fillna(df['Crop'].mode()[0],inplace=True)
print(df['Crop'].head())
df.isnull().sum()
```
